```python
from ultralytics import YOLO
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# Load the trained model
model = YOLO('/content/runs/detect/train2/weights/best.pt')

# Perform prediction on the uploaded image with a low confidence threshold
results = model.predict(source=image_path, conf=0.2)

# Check if there are any detections
if results[0].boxes is not None and len(results[0].boxes) > 0:
    # Open the original image
    predicted_image = Image.open(image_path)
    draw = ImageDraw.Draw(predicted_image)

    # Loop through each detection box
    for box in results[0].boxes:
        # Get the coordinates and confidence score of each box
        x1, y1, x2, y2 = box.xyxy[0]
        confidence = box.conf[0]

        # Draw the bounding box on the image
        draw.rectangle(((x1, y1), (x2, y2)), outline="red", width=3)
        draw.text((x1, y1), f"{confidence:.2f}", fill="red")

    # Display the image with bounding boxes
    plt.imshow(predicted_image)
    plt.axis('off')
    plt.show()

else:
    print("No detections found at the specified confidence threshold.")
```
